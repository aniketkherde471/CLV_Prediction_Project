```python
numerical_features = dataD.select_dtypes(include=np.number).columns
categorical_features = [i for i in dataD.columns if i not in numerical_features]
```

```python
categorical_features
```

```python
data_numerical_filled = dataD[numerical_features].fillna(dataD[numerical_features].mean())
data_categorical_filled = dataD[categorical_features].fillna(dataD[categorical_features].mode())
data_categorical_filled
```

```python
data_categorical_filled_ohe = pd.get_dummies(data_categorical_filled)
```

```python
final_data = pd.concat([data_numerical_filled, data_categorical_filled_ohe], axis = 1)
```

```python
X = final_data.drop(["SalePrice"], axis = 1) #feature
y = final_data["SalePrice"] #target
```

```python
#ANalysis
#Feature ENgineering
#Training
#Evaluation
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xval, ytrain, yval = train_test_split(X, y, test_size = 0.2, random_state = 54)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(Xtrain, ytrain)
```

```python
pred = lr.predict(Xval)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
(mean_absolute_error(pred, yval))
```
